### 5.1.3.2 Transfer Functions in Series

https://orb.binghamton.edu/cgi/viewcontent.cgi?filename=10&article=1002&context=electrical_fac&type=additional

Page 383

```python
import control as control
import numpy as np
from numpy.linalg import matrix_rank, inv
from numpy import matmul, around

T = np.linspace(0, 9, 10)
U = np.ones((1, 10))
sys = control.tf([2, 1], [1, 2, 3], 1)
T, Y = control.forced_response(sys, T, U)

h = control.markov(Y, U)[0]

def H(nr, nc, dr=0, dc=0):
    return np.array([[round(h[dc + dr * dc + j + 1]) for j in range(i, i + nc)] for i in range(nr)])

# page 385 
# """If n is not known, we can successively test the rank of the Hankel
# matrix as we add rows and columns to it, until it appears to stop growing in rank.""""
def rank():
    i = 1
    while matrix_rank(H(i, i)) != matrix_rank(H(i + 1, i + 1)):
        i = i + 1
    return i

n = rank()
print(n)
M = H(2, 2)
MA = H(2, 2, 0, 1)
MC = H(1, 2)
MB = H(2, 1)

print(f'M={M}\nMA={MA}\nMB={MB}\nMC={MC}')

A = around(matmul(MA, inv(M)))
b = around(MB)
c = around(matmul(MC, inv(M)))
d = round(h[0])

print(f'A={A}\nb={b}\nc={c}\nd={d}')

isys = control.ss2tf(A, b, c, d)

print(f'isys={isys}')
```
